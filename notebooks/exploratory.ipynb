```python
from torch.utils.data import DataLoader, dataloader
from  matplotlib import pyplot as plt
from data.data_loader import *
import albumentations as A
from albumentations.pytorch import ToTensorV2
from torch.utils.data import ConcatDataset
from utils.visualization import analyze_class_distribution

plt.style.use('default')
DATASET1_DIR='../data/dataset/dataset1'
DATASET2_DIR='../data/dataset/dataset2'
```

```python
albu_transform = A.Compose([
    ToTensorV2()
])
dataset1 = FootballDataset1(
    os.path.join(DATASET1_DIR,'images'),
    os.path.join(DATASET1_DIR,'masks'),
    transform=albu_transform
)
dataloader1 = DataLoader(dataset1, batch_size=4, shuffle=True)
```

```python

dataset2 = FootballDataset2(
    os.path.join(DATASET2_DIR,'images'),
    transform=albu_transform
)
dataloader2 = DataLoader(dataset2, batch_size=10, shuffle=True)
```

```python
combined =ConcatDataset([dataset1, dataset2])
dataloader = DataLoader(combined, batch_size=4, shuffle=True)
```

```python

def show_examples(dataloader):
    images, masks = next(iter(dataloader))
    fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(16, 8), layout='constrained')
    fig.suptitle('Dataset: Images and Unified Masks', fontsize=16)
    print(f"Unique integers in this batch: {torch.unique(masks).numpy()}")
    for i in range(4):
        img_np = images[i].permute(1, 2, 0).numpy()
        mask_np = masks[i].numpy()
        axes[0, i].imshow(img_np)
        axes[0, i].set_title(f"Image {i+1}")
        axes[0, i].axis('off')
        im = axes[1, i].imshow(mask_np, cmap='tab10', vmin=0, vmax=8)
        axes[1, i].set_title(f"Mask {i+1}")
        axes[1, i].axis('off')

    cbar = fig.colorbar(im, ax=axes.ravel().tolist(), orientation='vertical', fraction=0.02, pad=0.04)
    cbar.set_label('Class Index (0-8)')
    plt.show()
```

```python
show_examples(dataloader)
```
